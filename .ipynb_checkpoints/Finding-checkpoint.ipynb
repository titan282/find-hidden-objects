# 1. Import Library

```python
import numpy as np
import cv2
import imutils
%matplotlib inline
import matplotlib.pyplot as plt
```

# 2. Read  Image Assignment and Template

```python
# Read SOURCE_IMG
image_assignment = cv2.imread('./Assets/source-exam.jpg') # image assignment
image_origin = cv2.imread('./Assets/origin.jpg') # image origin crop t assignment

# Display image-source
plt.figure(figsize=(15,15))
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(image_assignment,cv2.COLOR_BGR2RGB))
plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(image_origin,cv2.COLOR_BGR2RGB))
```

```python
#Read Template Image
template=[]
template.append(cv2.imread('./Assets/template-strawberry.jpg',0)) # template image
template.append(cv2.imread('./Assets/template-bone.jpg',0))
template.append(cv2.imread('./Assets/template-bird.jpg',0))
template.append(cv2.imread('./Assets/template-bow.jpg',0))
template.append(cv2.imread('./Assets/template-cheese.jpg',0))
template.append(cv2.imread('./Assets/template-cloud.jpg',0))
template.append(cv2.imread('./Assets/template-icecream.jpg',0))
template.append(cv2.imread('./Assets/template-bug.jpg',0))
template.append(cv2.imread ('./Assets/template-butterfly.jpg',0))
template.append(cv2.imread('./Assets/template-icecream2.jpg',0))
template.append(cv2.imread('./Assets/template-sausage.jpg',0))
template.append(cv2.imread('./Assets/template-tennis.jpg',0))
for i in range(0,12):
    plt.subplot(3, 4, i+1)
    plt.imshow(template[i],cmap='gray')
    
```

# 3. Finding Template in Origin

```python
origin_gray = cv2.cvtColor(image_origin, cv2.COLOR_BGR2GRAY)
threshold = 0.6

#Hm v Rect v putText ca nhng template cn tm 
def drawRect(template,name): 
    w, h = template.shape[::-1] # Ly width v height
    for scale in np.linspace(0.2, 1.0, 20)[::-1]: # Ly 20 gi tr scale chia u t 1
        resized = imutils.resize(template, width = int(template.shape[1] * scale))
        w, h = resized.shape[::-1] # Ly width v height sau khi resized
        res = cv2.matchTemplate(origin_gray,resized,cv2.TM_CCOEFF_NORMED) # Find Correlation Coefficient Template Matching

        loc = np.where( res >= threshold) #Finding Location of template in origin
        if len(list(zip(*loc[::-1]))) > 0:
            break
    
    #Draw finding result
    if loc and len(list(zip(*loc[::-1]))) > 0:
        for pt in zip(*loc[::-1]):
            cv2.rectangle(image_origin, pt, (pt[0] + w, pt[1] + h), (0,0,255), 2)
            cv2.putText(image_origin,name,(pt[0]-5,pt[1]-10),cv2.FONT_HERSHEY_SIMPLEX,1.5,(0,0,255), 2)
# Execute Function            
# for i in range(0,12):
#     drawRect(template[i],'temp'+str(i+1))
# C th vng lp ny cho nhanh tuy nhin  tng minh th em s gn label cho n
drawRect(template[0],'strawberry')
drawRect(template[1],'bone')
drawRect(template[2],'bird')
drawRect(template[3],'bow')
drawRect(template[4],'cheese')
drawRect(template[5],'cloud')
drawRect(template[6],'icecream')
drawRect(template[7],'bug')
drawRect(template[8],'butterfly')
drawRect(template[9],'icecream')
drawRect(template[10],'sausage')
drawRect(template[11],'tennis')

#Display Output
plt.figure(figsize=(15,15))
plt.imshow(cv2.cvtColor(image_origin,cv2.COLOR_BGR2RGB))

#Running time hi lu c  khong 20s, c i t 
```
